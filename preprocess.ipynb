```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("human_vital_signs_dataset_2024.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
DROP_COLS = ['Patient ID', 'Timestamp']
df.drop(columns=DROP_COLS, inplace=True)
```

```python
df['Risk Category'] = (
    df['Risk Category']
    .astype(str)
    .str.lower()
    .str.strip()
)

df['Risk Category'] = df['Risk Category'].map({
    'low risk': 0,
    'high risk': 1
})
```

```python
df['Gender'] = (
    df['Gender']
    .astype(str)
    .str.lower()
    .str.strip()
)

df['Gender'] = df['Gender'].map({
    'male': 0,
    'female': 1
})
```

```python
print(df['Gender'].unique())
```

```python
num_cols = df.columns.drop('Risk Category')

df[num_cols] = df[num_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df = df[
    df['Heart Rate'].between(40, 180) &
    df['Oxygen Saturation'].between(70, 100) &
    df['Body Temperature'].between(34, 42)
]
```

```python
print(df.dtypes)
print(df.isnull().sum().sum())
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
